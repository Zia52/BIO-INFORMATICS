```python
def exon_chaining(exons):
    """
    Finds the maximum weight chain of non-overlapping exons.

    :param exons: List of tuples representing exons: (start, end, weight)
    :return: Tuple containing (max_weight, list_of_chained_exons)
    """
    if not exons:
        return 0, []

    # Sort exons by their starting coordinates
    sorted_exons = sorted(exons, key=lambda x: x[0])
    n = len(sorted_exons)

    # dp[i] will store the maximum weight of a valid chain ending at sorted_exons[i]
    dp = [0] * n

    # parent[i] tracks the index of the previous exon in the optimal chain
    parent = [-1] * n

    for i in range(n):
        # Base weight of the current exon
        dp[i] = sorted_exons[i][2]

        # Check all previous exons to find valid, non-overlapping chains
        for j in range(i):
            # Condition for non-overlap: start of current exon >= end of previous exon
            if sorted_exons[i][0] >= sorted_exons[j][1]:
                if dp[j] + sorted_exons[i][2] > dp[i]:
                    dp[i] = dp[j] + sorted_exons[i][2]
                    parent[i] = j

    # Find the exon that yields the maximum weight overall
    max_weight_idx = dp.index(max(dp))

    # Reconstruct the optimal chain by backtracking through parent pointers
    chain = []
    curr = max_weight_idx
    while curr != -1:
        chain.append(sorted_exons[curr])
        curr = parent[curr]

    chain.reverse()  # Reverse to get the correct chronological order

    return dp[max_weight_idx], chain

# --- Example Usage ---
if __name__ == "__main__":
    # Exon format: (start, end, weight)
    # Exons must not overlap, e.g., next start must be >= previous end
    candidate_exons = [
        (1, 3, 5),
        (2, 5, 4),
        (4, 7, 8),
        (6, 9, 2),
        (8, 10, 6)
    ]

    max_score, optimal_chain = exon_chaining(candidate_exons)
    print(f"Maximum Total Weight: {max_score}")
    print(f"Optimal Exon Chain: {optimal_chain}")
```
