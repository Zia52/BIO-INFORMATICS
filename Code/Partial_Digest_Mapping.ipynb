```python
def partial_digest(L):
    """
    Main function to solve the Partial Digest Problem.
    Input: L - a list (multiset) of pairwise distances.
    Output: A list of sorted coordinates of the restriction sites.
    """
    if not L:
        return []

    # Clean the input by making a copy and working with integers
    L = list(L)

    # The maximum distance in L represents the total width of the DNA molecule
    width = max(L)
    L.remove(width)

    # Initialize the solution set with the boundary points
    X = [0, width]

    solutions = []
    place(L, X, width, solutions)

    # Return unique, sorted solutions
    unique_solutions = sorted(list(set(tuple(sorted(sol)) for sol in solutions)))
    return [list(sol) for sol in unique_solutions]


def place(L, X, width, solutions):
    """
    Recursive backtracking function using the Branch and Bound approach.
    """
    # Base case: if the multiset L is empty, a complete configuration X is found
    if not L:
        solutions.append(list(X))
        return

    # Find the maximum remaining distance in L
    y = max(L)

    # Alternative 1: Test placing a restriction site at coordinate x = y
    if is_valid_site(y, X, L):
        # Action: Add site and remove corresponding pairwise distances from L
        X.append(y)
        removed_distances = remove_distances(y, X, L)

        # Recurse
        place(L, X, width, solutions)

        # Backtrack: Restore X and L
        X.remove(y)
        L.extend(removed_distances)

    # Alternative 2: Test placing a restriction site at coordinate x = width - y
    alternative_site = width - y
    if is_valid_site(alternative_site, X, L):
        # Action: Add site and remove corresponding pairwise distances from L
        X.append(alternative_site)
        removed_distances = remove_distances(alternative_site, X, L)

        # Recurse
        place(L, X, width, solutions)

        # Backtrack: Restore X and L
        X.remove(alternative_site)
        L.extend(removed_distances)


def is_valid_site(y, X, L):
    """
    Checks if all distances between target site 'y' and existing sites in 'X' exist in 'L'.
    """
    # Count the required frequencies of each distance
    required_distances = {}
    for xi in X:
        dist = abs(y - xi)
        required_distances[dist] = required_distances.get(dist, 0) + 1

    # Validate against available elements in L
    for dist, count in required_distances.items():
        if L.count(dist) < count:
            return False
    return True


def remove_distances(y, X, L):
    """
    Removes the pairwise distances between 'y' and all existing items in 'X' from 'L'.
    Returns the removed items to facilitate backtracking.
    """
    removed = []
    for xi in X:
        if xi != y:  # Skip distance with itself (which is 0)
            dist = abs(y - xi)
            L.remove(dist)
            removed.append(dist)
    return removed


# --- Example Execution ---
if __name__ == "__main__":
    # Given multiset of pairwise distances
    pairwise_distances = [2, 2, 3, 3, 4, 5, 6, 7, 8, 10]

    print(f"Input Pairwise Distances (L): {pairwise_distances}")
    results = partial_digest(pairwise_distances)

    print("\nReconstructed DNA Restriction Maps (X):")
    for idx, solution in enumerate(results, start=1):
        print(f"Solution {idx}: {solution}")
```
