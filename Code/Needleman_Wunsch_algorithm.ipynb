```python
def needleman_wunsch(seq1, seq2, match=1, mismatch=-1, gap=-1):
    """
    Computes the optimal global alignment of two sequences
    using the Needleman-Wunsch algorithm.
    """
    n = len(seq1)
    m = len(seq2)

    # 1. Initialize the scoring matrix with zeros
    matrix = [[0] * (m + 1) for _ in range(n + 1)]

    # Fill the base cases for the first row and column
    for i in range(n + 1):
        matrix[i][0] = i * gap
    for j in range(m + 1):
        matrix[0][j] = j * gap

    # 2. Fill the rest of the matrix using the recurrence relation
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            # Calculate match/mismatch score
            if seq1[i - 1] == seq2[j - 1]:
                score_diag = matrix[i - 1][j - 1] + match
            else:
                score_diag = matrix[i - 1][j - 1] + mismatch

            score_up = matrix[i - 1][j] + gap
            score_left = matrix[i][j - 1] + gap

            # Record the maximum score
            matrix[i][j] = max(score_diag, score_up, score_left)

    # 3. Traceback from the bottom-right corner to find the alignment
    align1 = []
    align2 = []
    i, j = n, m

    while i > 0 or j > 0:
        if i > 0 and j > 0:
            # Check if the score came from a match/mismatch (diagonal)
            score_match = match if seq1[i - 1] == seq2[j - 1] else mismatch
            if matrix[i][j] == matrix[i - 1][j - 1] + score_match:
                align1.append(seq1[i - 1])
                align2.append(seq2[j - 1])
                i -= 1
                j -= 1
                continue

        if i > 0:
            # Check if the score came from a deletion (upward move)
            if matrix[i][j] == matrix[i - 1][j] + gap:
                align1.append(seq1[i - 1])
                align2.append("-")
                i -= 1
                continue

        if j > 0:
            # Check if the score came from an insertion (leftward move)
            if matrix[i][j] == matrix[i][j - 1] + gap:
                align1.append("-")
                align2.append(seq2[j - 1])
                j -= 1
                continue

    # Reverse the aligned sequences since they were collected backwards
    aligned_seq1 = "".join(reversed(align1))
    aligned_seq2 = "".join(reversed(align2))
    alignment_score = matrix[n][m]

    return aligned_seq1, aligned_seq2, alignment_score


# --- Example Usage ---
if __name__ == "__main__":
    sequence_a = "GATTACA"
    sequence_b = "GCATGCU"

    # Run alignment with standard DNA parameters
    res1, res2, score = needleman_wunsch(sequence_a, sequence_b, match=1, mismatch=-1, gap=-1)

    print(f"Alignment Score: {score}")
    print(f"Sequence 1: {res1}")
    print(f"Sequence 2: {res2}")
```
